첫 페이지 댓글 수집

````python
[response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('textOriginal') for i in range(len(response.get('items')))]
````

````python
def get_comments_from_youtube_api_result(response):
    return [response.get('items')[i].get('snippet').get('topLevelComment').get('snippet').get('textOriginal') for i in range(len(response.get('items')))]
````

전체 댓글 수집

````python
all_comments = []

#첫번째 페이지
response = (
    youtube
    .commentThreads()
    .list(
      part='snippet,replies', 
      videoId='uRFEhFIsyXY', 
      maxResults=100
      )
    .execute()
    )
all_comments.extend(get_comments_from_youtube_api_result(response))

#첫페이지 포함안됨  
while response.get('nextPageToken') :
    response = (
      youtube
      .commentThreads()
      .list(
        part='snippet,replies',
        videoId='uRFEhFIsyXY',
        pageToken=response.get('nextPageToken'),
        maxResults=100
        )
      .execute()
    )
    all_comments.extend(get_comments_from_youtube_api_result(response))
````
